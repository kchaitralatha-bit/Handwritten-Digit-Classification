<h2 style="color:black" align="center">Handwritten digits classification using neural network</h2>

In this coding part  we will classify handwritten digits using a simple neural network which has only input and output layers. We will than add a hidden layer and see how the performance of the model improves

```python
import tensorflow as tf
from tensorflow import keras
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
(X_train, y_train) , (X_test, y_test) = keras.datasets.mnist.load_data()
```

```python
y_train[0]
```

```python
X_train = X_train / 255
X_test = X_test / 255
```

```python
X_train[0]
```

```python
X_train_flattened = X_train.reshape(len(X_train), 28*28)
X_test_flattened = X_test.reshape(len(X_test), 28*28)
```

```python
X_train_flattened.shape
```

```python
X_train_flattened[0]
```

<h3 style='color:black'>Very simple neural network with no hidden layers</h3>

```python
model = keras.Sequential([
    keras.layers.Dense(10, input_shape=(784,), activation='sigmoid')
])

model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

model.fit(X_train_flattened, y_train, epochs=5)
```

```python
model.evaluate(X_test_flattened, y_test)
```

```python
y_predicted = model.predict(X_test_flattened)
y_predicted[0]
```

```python
plt.matshow(X_test[0])
```

**np.argmax finds a maximum element  from an array and returns the index of it**

```python
np.argmax(y_predicted[0])
```

```python
y_predicted_labels = [np.argmax(i) for i in y_predicted]
```

```python
y_predicted_labels[:5]
```

```python
cm = tf.math.confusion_matrix(labels=y_test,predictions=y_predicted_labels)
cm
```

```python
import seaborn as sn
plt.figure(figsize = (10,7))
sn.heatmap(cm, annot=True, fmt='d')
plt.xlabel('Predicted')
plt.ylabel('Truth')
```

<h3 style='color:black'>Using hidden layer</h3>

```python
model = keras.Sequential([
    keras.layers.Dense(100, input_shape=(784,), activation='relu'),
    keras.layers.Dense(10, activation='sigmoid')
])

model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

model.fit(X_train_flattened, y_train, epochs=5)
```

```python
model.evaluate(X_test_flattened,y_test)
```

```python
y_predicted = model.predict(X_test_flattened)
y_predicted_labels = [np.argmax(i) for i in y_predicted]
cm = tf.math.confusion_matrix(labels=y_test,predictions=y_predicted_labels)

plt.figure(figsize = (10,7))
sn.heatmap(cm, annot=True, fmt='d')
plt.xlabel('Predicted')
plt.ylabel('Truth')
```

<h3 style='color:black'>Using Flatten layer so that we don't have to call .reshape on input dataset</h3>

```python
model = keras.Sequential([
    keras.layers.Flatten(input_shape=(28, 28)),
    keras.layers.Dense(100, activation='relu'),
    keras.layers.Dense(10, activation='sigmoid')
])

model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

model.fit(X_train, y_train, epochs=10)
```

```python
model.evaluate(X_test,y_test)
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data for chart
models = ['Model 1: Dense with Sigmoid', 
          'Model 2: Two Layers with ReLU', 
          'Model 3: Flatten + ReLU']
train_accuracy = [92.62, 98.12, 99.17]  # Training Accuracy in percentages
test_accuracy = [92.57, 97.16, 97.79]  # Test Accuracy in percentages
train_loss = [0.2677, 0.0618, 0.0264]  # Training Loss
test_loss = [0.2670, 0.0966, 0.0813]  # Test Loss

# Bar width
bar_width = 0.2

# Positions for bars
r1 = np.arange(len(models))
r2 = [x + bar_width for x in r1]

# Create subplots
fig, axs = plt.subplots(1, 2, figsize=(15, 6))

# Accuracy Chart
bars1 = axs[0].bar(r1, train_accuracy, color='blue', width=bar_width, label='Train Accuracy')
bars2 = axs[0].bar(r2, test_accuracy, color='orange', width=bar_width, label='Test Accuracy')
axs[0].set_title('Model Accuracy')
axs[0].set_xticks([r + bar_width / 2 for r in r1])
axs[0].set_xticklabels(models, rotation=15, ha='right')
axs[0].set_ylabel('Accuracy (%)')
axs[0].legend()
axs[0].grid(axis='y', linestyle='--', alpha=0.7)

# Display values on bars for accuracy
for bar in bars1:
    axs[0].text(bar.get_x() + bar.get_width() / 2, bar.get_height(),
                f'{bar.get_height():.2f}', ha='center', va='bottom', fontsize=9)
for bar in bars2:
    axs[0].text(bar.get_x() + bar.get_width() / 2, bar.get_height(),
                f'{bar.get_height():.2f}', ha='center', va='bottom', fontsize=9)

# Loss Chart
bars3 = axs[1].bar(r1, train_loss, color='blue', width=bar_width, label='Train Loss')
bars4 = axs[1].bar(r2, test_loss, color='orange', width=bar_width, label='Test Loss')
axs[1].set_title('Model Loss')
axs[1].set_xticks([r + bar_width / 2 for r in r1])
axs[1].set_xticklabels(models, rotation=15, ha='right')
axs[1].set_ylabel('Loss')
axs[1].legend()
axs[1].grid(axis='y', linestyle='--', alpha=0.7)

# Display values on bars for loss
for bar in bars3:
    axs[1].text(bar.get_x() + bar.get_width() / 2, bar.get_height(),
                f'{bar.get_height():.4f}', ha='center', va='bottom', fontsize=9)
for bar in bars4:
    axs[1].text(bar.get_x() + bar.get_width() / 2, bar.get_height(),
                f'{bar.get_height():.4f}', ha='center', va='bottom', fontsize=9)

# Adjust layout and show plot
plt.tight_layout()
plt.show()
```
